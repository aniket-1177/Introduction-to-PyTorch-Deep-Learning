## tensor maths operations

- add, subtract, multiply, divide, remainders, exponents
- shorthand, adn longhand
- reassingment

```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
#addition

tensor_a + tensor_b
```

```python
#addition long hand

torch.add(tensor_a, tensor_b)
```

```python
# subtraction   
tensor_a - tensor_b
```

```python
torch.subtract(tensor_a, tensor_b)
```

```python
#multiplication
tensor_a * tensor_b
```

```python
torch.multiply(tensor_a, tensor_b)
```

```python
#division
tensor_a / tensor_b
```

```python
torch.divide(tensor_a, tensor_b)
```

```python
#remainder
tensor_a % tensor_b
```

```python
torch.remainder(tensor_a, tensor_b)
```

```python
#exponents / power

torch.pow(tensor_a, tensor_b)
```

```python
tensor_a ** tensor_b
```

```python
tensor_a.add(tensor_b)
```

```python
#reassignment : done using _ symbol (underscore symbol)

tensor_a + tensor_b
```

```python
tensor_a
```

```python
# tensor_a = tensor_a + tensor_b

tensor_a.add_(tensor_b)
```
