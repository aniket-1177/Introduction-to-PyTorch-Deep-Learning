## Difference between reshape() and view()

https://stackoverflow.com/questions/49643225/whats-the-difference-between-reshape-and-view-in-pytorch

https://mr-amit.medium.com/pytorch-view-vs-reshape-a-practical-guide-74bf31eec210


The primary difference between view and reshape in PyTorch is how they handle underlying memory and tensor contiguity.

- view strictly reinterprets the existing memory layout and never copies data, meaning it only works on contiguous tensors.

- reshape acts as a safer, more flexible wrapper; it returns a view if possible, but automatically allocates a new copy of the data if the tensor is non-contiguous.

- A tensor is contiguous when its elements are stored sequentially in a single, uninterrupted block of computer memory.
